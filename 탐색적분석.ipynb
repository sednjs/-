## 25) 최대 월 납입 보험료(MAX_MON_PREM)

```python
# 연령대별(AGEGROUP) 최대 월 납입 보험료
print(df[df['AGEGROUP']=="20"]['MAX_MON_PREM'].shape,
df[df['AGEGROUP']=="30"]['MAX_MON_PREM'].shape,
df[df['AGEGROUP']=="40"]['MAX_MON_PREM'].shape,
df[df['AGEGROUP']=="50"]['MAX_MON_PREM'].shape,
df[df['AGEGROUP']=="60"]['MAX_MON_PREM'].shape,
df[df['AGEGROUP']=="70"]['MAX_MON_PREM'].shape)
```

```python
# 제일 적은 수인 20대의수를 기준으로 랜덤으로 sampling 
twe= df[df['AGEGROUP']=="20"]['MAX_MON_PREM'].sample(n=305,random_state=0).to_list()
thir = df[df['AGEGROUP']=="30"]['MAX_MON_PREM'].sample(n=305,random_state=0).to_list()
four = df[df['AGEGROUP']=="40"]['MAX_MON_PREM'].sample(n=305,random_state=0).to_list()
five = df[df['AGEGROUP']=="50"]['MAX_MON_PREM'].sample(n=305,random_state=0).to_list()
six = df[df['AGEGROUP']=="60"]['MAX_MON_PREM'].sample(n=305,random_state=0).to_list()
sev = df[df['AGEGROUP']=="70"]['MAX_MON_PREM'].sample(n=305,random_state=0).to_list()
```

```python
# 가설: 최대 월 납입 보험료는 연령대 별로 다를 것이다.
string_list = ['twenty','thirty','fourty','fifty','sixty','seventy']
int_list = [twe,thir,four,five,six,sev]
dictionary = dict(zip(string_list, int_list))

df5=pd.DataFrame(dictionary)

# 일원분산분석 
# 귀무가설: 최대 월 납입 보험료와 연령의 분포가 같다 
# 대립가설: 최대 월 납입 보험료의 평균의 분포에 차이가 있다. 

f_result= stats.f_oneway(df5['twenty'],df5['thirty'],df5['fourty'],df5['fifty'],df5['sixty'],df5['seventy'])
f,p = f_result.statistic.round(3),f_result.pvalue.round(3)
# f score, p-value 
print(f,p)
```

#### p value가 0.05 보다 크므로 두 집단의 분포에는 차이가 없다. (귀무가설 채택, 연령간의 차이가 없음 )
