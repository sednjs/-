```python
list_tot=[]
list_fm=[]
target_fm=[]

for x in df_ins['ACTL_FMLY_NUM']:
    list_tot.append(x)
    
for y in df_ins['CUST_FMLY_NUM']:
    list_fm.append(y)

# len(list_tot)  
# len( list_fm)
for i in range(len(list_tot)):
    target_fm.append(list_tot[i]-list_fm[i])
```
